```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,10,100)
y = x **2
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
axes1.plot(x,y)
axes1.set_title('Larger plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
fig
```

```python
fig,ax = plt.subplots(nrows = 1, ncols = 2,figsize = (8,2))
plt.tight_layout()
ax[0].plot(x,x**2,label = 'X suqared')
ax[0].plot(x,x**3,label = 'X cubed')
ax[0].legend(loc = 0)
```
